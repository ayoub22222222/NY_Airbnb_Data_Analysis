# Step 4: Interactive Visualization and Analytical Questions

```python
filtre_data = df[df['price'] <= 10000]
```

 #### Visualizing Price Distribution Across Room Types

```python
# Define price bins
bins = [0, 100, 200, 400, 800, 1000,10000]
bin_labels = ['0-100', '100-200', '200-400', '400-800', '800-1000','1000-10000']

# Add a new column for binned prices
filtre_data['price_bin'] = pd.cut(filtre_data['price'], bins=bins, labels=bin_labels, include_lowest=True)

# Create a binned bar plot for price distribution by room type
plt.figure(figsize=(12,8))
sns.countplot(x='price_bin', hue='room_type', data=filtre_data, palette='coolwarm')

# Add a title and axis labels
plt.title('Price Distribution by Room Type (Binned Price Ranges)', fontsize=16)
plt.xlabel('Price Range (in USD)', fontsize=14)
plt.ylabel('Number of Listings', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.legend(title='Room Type', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```

# Relationship Between Price and Number of Review

```python
#  create a scatter plot to visualize the relationship between price and number of reviews

plt.figure(figsize=(10, 6))
sns.scatterplot(x='price', y='number_of_reviews', data=filtre_data, palette='coolwarm')
plt.title("visualize the relationship between price and number of reveiws")
plt.xlabel("price (in USD)")
plt.xlabel("Number of Reviews")
plt.show()
```

# Avg Price of AirBnb Listings by Neighborhood Group
💡 **Objective:** Find out which neighborhood groups have the highest average listing prices.


```python
new_df = df.groupby('neighbourhood_group')['price'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
new_df.plot(kind='bar', color='teal')
plt.title("Avg Price Listing By Neighbourhood Group")
plt.xlabel("Neighbourhood Group")
plt.ylabel("Average Price (in USD)")
plt.show()
```

## Which room types have the highest average number of reviews?


```python
grouped_room_type = filtre_data.groupby('room_type')
ploted_result = grouped_room_type['number_of_reviews'].mean().sort_values(ascending=False)
```
